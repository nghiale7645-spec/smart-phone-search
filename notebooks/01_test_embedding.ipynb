```python
import pandas as pd
from sentence_transformers import SentenceTransformer

# 1. Tạo dữ liệu điện thoại mẫu
data = {
    "id": ["p1", "p2", "p3"],
    "name": ["iPhone 15 Pro Max", "Samsung Galaxy S24 Ultra", "Xiaomi Redmi Note 13"],
    "description": [
        "Điện thoại cao cấp mượt mà, chip xử lý siêu mạnh, camera chụp ảnh quay phim sắc nét, thiết kế titan.",
        "Điện thoại màn hình lớn, có bút cảm ứng tiện lợi cho công việc, camera zoom xa, pin dung lượng cao.",
        "Điện thoại giá rẻ bình dân, pin cực trâu dùng cả ngày, màn hình to xem phim tốt, chơi game nhẹ."
    ]
}
df = pd.DataFrame(data)
print("--- Dữ liệu gốc ---")
display(df)

# 2. Tải mô hình AI đa ngôn ngữ (Hỗ trợ tiếng Việt)
print("\nĐang tải mô hình AI... (Lần đầu chạy sẽ tải khoảng hơn 400MB)")
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# 3. Chuyển đổi phần 'description' thành Vector số
embeddings = model.encode(df['description'].tolist())

print("\n--- Kết quả Embedding ---")
print(f"Đã tạo thành công vector cho {len(embeddings)} điện thoại.")
print(f"Mỗi câu mô tả đã biến thành một mảng số học gồm {len(embeddings[0])} chiều.")
print(f"Ví dụ một vài số đầu tiên của Vector số 1: {embeddings[0][:5]}")
```
